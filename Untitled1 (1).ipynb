```python
# Import libraries
import numpy as np
import pandas as pd

# Generate dataset
np.random.seed(42)
num_rows = 200
customer_ids = np.random.choice([f"CUST{str(i).zfill(4)}" for i in range(1, 51)], num_rows)
ages = np.random.normal(35, 10, num_rows).astype(float)
ages[np.random.choice(num_rows, 10, replace=False)] = np.nan
genders = np.random.choice(["Male", "Female", "Other"], num_rows)
purchase_amounts = np.random.normal(500, 200, num_rows).astype(int)
purchase_amounts[np.random.choice(num_rows, 5, replace=False)] = [5000, 7000, 6500, 8000, 9000]
product_categories = np.random.choice(["Electronics", "Clothing", "Home & Kitchen", "Beauty", "Sports"], num_rows)
transaction_dates = pd.date_range(start="2024-01-01", periods=num_rows, freq="D")
payment_methods = np.random.choice(["Credit Card", "Debit Card", "UPI", "Net Banking", "Cash on Delivery"], num_rows)
discount_applied = np.random.randint(0, 50, num_rows).astype(float)
discount_applied[np.random.choice(num_rows, 15, replace=False)] = np.nan
order_statuses = np.random.choice(["Completed", "Pending", "Cancelled", "Returned"], num_rows)

df = pd.DataFrame({
    "Customer ID": customer_ids,
    "Age": ages,
    "Gender": genders,
    "Purchase Amount": purchase_amounts,
    "Product Category": product_categories,
    "Transaction Date": transaction_dates,
    "Payment Method": payment_methods,
    "Discount Applied": discount_applied,
    "Order Status": order_statuses
})

# Add duplicate rows and shuffle dataset
df = pd.concat([df, df.iloc[:5]], ignore_index=True)
df = df.sample(frac=1, random_state=42).reset_index(drop=True)

# Display first few rows
df.head()
```

```python
# Handle missing values
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Discount Applied'].fillna(df['Discount Applied'].mean(), inplace=True)

# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Detect and treat outliers in 'Purchase Amount'
Q1 = df['Purchase Amount'].quantile(0.25)
Q3 = df['Purchase Amount'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

median_value = df['Purchase Amount'].median()
df.loc[(df['Purchase Amount'] < lower_bound) | (df['Purchase Amount'] > upper_bound), 'Purchase Amount'] = median_value

# Standardize categorical columns
categorical_cols = ['Gender', 'Product Category', 'Payment Method', 'Order Status']
for col in categorical_cols:
    df[col] = pd.Categorical(df[col])

# Display cleaned dataset
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Histogram for Age distribution
sns.histplot(df['Age'], kde=True).set_title('Age Distribution')
plt.show()

# Box plot for Purchase Amount distribution
sns.boxplot(x=df['Purchase Amount']).set_title('Purchase Amount Distribution')
plt.show()
```

```python
# Scatter plot: Age vs Purchase Amount
sns.scatterplot(x='Age', y='Purchase Amount', data=df).set_title('Age vs Purchase Amount')
plt.show()

# Bar plot: Product Category vs Average Purchase Amount
sns.barplot(x='Product Category', y='Purchase Amount', data=df).set_title('Average Purchase Amount by Product Category')
plt.show()
```

```python
# Pair plot for numerical variables
sns.pairplot(df[['Age', 'Purchase Amount', 'Discount Applied']])
plt.show()

# Heatmap for correlation matrix of numerical variables
correlation_matrix = df[['Age', 'Purchase Amount', 'Discount Applied']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm').set_title('Correlation Matrix')
plt.show()
```
